```python
import numpy as np
import matplotlib.pyplot as plt
```

## Initial Value Problem 1

```python
h_1=0.1
y0=np.exp(1)
x_array=np.arange(0,1+h_1,h_1)
y_array_1=[]
y_array_1.append(y0)
for i in x_array[1:]:
    y1_new=y0/(1+9*h_1)
    y_array_1.append(y1_new)
    y0=y1_new
plt.plot(x_array,y_array_1)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## Initial Value Problem 2

$y_{j+1}=y_j+h*f(y_{j+1},x_{j+1})$ \
$y_{j+1}=y_j+h*(-20(y_{j+1}-x_{j+1})^2+2x_{j+1})$\
$20h y_{j+1}^2+(1-40hx_{j+1})y_{j+1}=y_j-20hx_{j+1}^2+2hx_{j+1}$

```python
h_2=0.01
y01=1/3
x_array=np.arange(0,1+h_2,h_2)
#print(x_array)
y_array_2=[]
y_array_2.append(y01)
for i in x_array[1:]:
    y2_new=np.roots([20*h_2,(1-40*h_2*i),-y01+20*h_2*i*i -2*h_2*i])
    y_array_2.append(y2_new[1])
    y01=y2_new[1]
plt.plot(x_array,y_array_2)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
