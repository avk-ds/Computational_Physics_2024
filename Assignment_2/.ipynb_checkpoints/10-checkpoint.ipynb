```python
import numpy as np
import matplotlib.pyplot as plt
from math import *


def runge_kutta_4th_order(func, time, value, step):
    k1 = step * func(time, value)
    k2 = step * func(time + step / 2, value + k1 / 2)
    k3 = step * func(time + step / 2, value + k2 / 2)
    k4 = step * func(time + step, value + k3)
    return value + (k1 + 2 * k2 + 2 * k3 + k4) / 6

def function(time, value):
    return (value**2 + value) / time

min_time = 1
max_time = 3
step = 0.001
values = [-2]
time_points = [min_time]
flag = 0 
while(flag == 0): 
    flag = 1
    if (time_points[-1] + step < max_time):
        flag = flag - 1
        next_value = runge_kutta_4th_order(function, time_points[-1], values[-1], step)
        values.append(next_value)
        time_points.append(time_points[-1] + step)
        next_value = runge_kutta_4th_order(function, time_points[-1], values[-1], step)
        prev_value = runge_kutta_4th_order(function, time_points[-2], values[-2], 2 * step)
        rho = 1
        if (np.abs(next_value - prev_value) != 0):
            rho = (10**(-4) * step * 30 / np.abs(next_value - prev_value))**(0.25)
        step = step * rho


step = max_time - time_points[-1]
values.append(runge_kutta_4th_order(function, time_points[-1], values[-1], step))
time_points.append(max_time)


def true_function(t):
    return -2 * t / (-1 + 2 * t)

error = np.linalg.norm(values - np.vectorize(true_function)(time_points), ord=inf)
print('The final error turns out to be = ', error)



plt.plot(time_points, values, 'red', label='Numerical')
for xc in time_points:
    plt.axvline(x=xc)

plt.plot(time_points, np.vectorize(true_function)(time_points), '--', color='green', label='Exact')


plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.show()
```
