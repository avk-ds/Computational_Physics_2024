## Defining the functions

```python
import numpy as np
import matplotlib.pyplot as plt
#We will break the 2nd order ode by substituting dy/dx=v.
#Defining the RHS of equation
def f_2(v,x,y):
    return (2*v-y+x*np.exp(x)-x)
def f_1(v):
    return v
def RK4(v_i,x_i,y_i):
    l1=h*f_2(v_i,x_i,y_i)
    k1=h*f_1(v_i)
    l2=h*f_2(v_i+(0.5*l1),x_i+0.5*h,(y_i+0.5*k1))
    k2=h*f_1(v_i+(0.5*l1))
    l3=h*f_2((v_i+0.5*l2),(x_i+0.5*h),y_i+(0.5*k2))
    k3=h*f_1(v_i+(0.5*l2))
    l4=h*f_2(v_i+(l3),x_i+h,(y_i+k3))
    k4=h*f_1(v_i+l3)
    y_f= (y_i+(1/6)*(k1+2*k2+2*k3+k4))
    v_f= (v_i +(1/6)*(l1+2*l2+2*l3+l4))
    return v_f, y_f
    
```

## Finding the solution

```python
v_old=0
y_old=0
h=0.1
x_array=np.arange(0,1+h,h)
y_array=[]
y_array.append(y_old)
for i in x_array[:-1]:
    v_old, y_old = RK4(v_old, i, y_old)
    y_array.append(y_old)
plt.plot(x_array, y_array)
plt.xlabel("x")
plt.ylabel("y")
plt.plot()
plt.show()
```
