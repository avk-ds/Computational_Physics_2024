# Equation 1

```python
import matplotlib.pyplot as plt
import scipy as sp
import numpy as np
def f1(t,y):
    return (t*np.exp(3*t)-2*y)
t1_span=[0,1]
t_1=np.arange(0,1,0.01)
y10=[0]
sol1=sp.integrate.solve_ivp(f1, t1_span, y10, t_eval=t_1)
plt.plot(sol1.t, sol1.y[0],label="scipy solution")
def f01(t):
    y1=((1/25)*np.exp(-2*t)*(1-np.exp(5*t)+5*np.exp(5*t)*t))
    return y1
plt.plot(t_1, f01(t_1), linestyle="--",label="Analytical Solution")
plt.xlabel("t")
plt.ylabel("y")
plt.title("Equation 1")
plt.legend()
plt.show()
```

# Equation 2

```python
def f2(t,y):
    return (1-pow((t-y),2))
t2_span=[2,3]
y20=[1]
t_2=np.arange(2,3,0.0001)
sol2=sp.integrate.solve_ivp(f2, t2_span, y20, t_eval=t_2)
plt.plot(sol2.t, sol2.y[0], label="Scipy solution")
def f02(t):
    y1=((1-3*t+t**2)/(t-3))
    return y1
plt.plot(t_2, f02(t_2), linestyle="--",label="Analytical Solution")
plt.xlabel("t")
plt.ylabel("y")
plt.title("Equation 2")
plt.legend()
plt.show()
```

# Equation 3

```python
def f3(t,y):
    return (1+(y/t))
t3_span=[1,2]
t_3=np.arange(1,2,0.0001)
y30=[2]
sol3=sp.integrate.solve_ivp(f3, t3_span, y30, t_eval=t_3)
plt.plot(sol3.t, sol3.y[0], label="Scipy Solution")
def f03(t):
    y1=(2*t+t*np.log(t))
    return y1
plt.plot(t_3, f03(t_3), linestyle="--",label="Analytical Solution")
plt.xlabel("t")
plt.ylabel("y")
plt.title("Equation 3")
plt.legend()
plt.show()
```

# Equation 4

```python
def f4(t,y):
    return (np.cos(2*t)+np.sin(3*t))
t4_span=[0,1]
t_4=np.arange(0,1,0.0001)
y40=[1]
sol4=sp.integrate.solve_ivp(f4, t4_span, y40, t_eval=t_4)
plt.plot(sol4.t, sol4.y[0], label="Scipy Solution")
plt.xlabel("t")
plt.ylabel("y")
def f04(t):
    y1=(1/6)*(8-2*np.cos(3*t)+3*np.sin(2*t))
    return y1
plt.plot(t_4, f04(t_4), linestyle="--",label="Analytical Solution")
plt.title("Equation 4")
plt.show()
```
