```python
import numpy as np
from scipy.integrate import solve_bvp
import matplotlib.pyplot as plt
```

## Equation 1

```python
def fun_1(x, y):
    return np.vstack((y[1], -np.exp(-2*y[0])))

def bc_1(ya, yb):
    return np.array([ya[0], yb[0]-np.log(2)])

x_1 = np.linspace(1, 2, 5)
y_1 = np.zeros((2, x_1.size))

sol_1 = solve_bvp(fun_1, bc_1, x_1, y_1)

x_plot1 = np.linspace(1, 2, 100)
y_plot1 = sol_1.sol(x_plot1)[0]

plt.plot(x_plot1, y_plot1, label="Scipy Solution")
def f01(t):
    y1=(np.log(t))
    return y1
plt.plot(x_plot1, f01(x_plot1), linestyle="--",label="Analytical Solution")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution of the 1st equation')
plt.legend()
plt.grid(True)
plt.show()
```

## Equation 2

```python
def fun_2(x, y):
    return np.vstack((y[1], y[1]*np.cos(x)-y[0]*np.log(y[0])))

def bc_2(ya, yb):
    return np.array([ya[0]-1, yb[0]-np.exp(1)])

x_2 = np.linspace(0, 3.14/2, 100)
y_2 = np.array([np.ones(x_2.shape), np.ones(x_2.shape)])

sol2 = solve_bvp(fun_2, bc_2, x_2, y_2)

x_plot2 = np.linspace(0, 3.14/2, 100)
y_plot2 = sol2.sol(x_plot2)[0]

# Plotting the solution
import matplotlib.pyplot as plt
plt.plot(x_plot2, y_plot2, label="Solution form Scipy")
def f02(t):
    y1=(np.exp(np.sin(t)))
    return y1
plt.plot(x_plot2, f02(x_plot2), linestyle="--",label="Analytical Solution")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution of the 2nd equation')
plt.legend()
plt.grid(True)
plt.show()
```

## Equation 3

```python
def fun_3(x, y):
    return np.vstack((y[1], -(2*pow(y[1],3)+pow(y[0],2)*y[1])/(np.cos(x))))

def bc_3(ya, yb):
    return np.array([ya[0]-pow(2,-0.25), yb[0]-(np.power(12,0.25))/2])

x_3 = np.linspace(3.14/4, 3.14/3, 5)
y_3 = np.zeros((2, x_3.size))

sol3 = solve_bvp(fun_3, bc_3, x_3, y_3)

x_plot3 = np.linspace(3.14/4, 3.14/3, 100)
y_plot3 = sol3.sol(x_plot3)[0]

def f03(t):
    y1=(np.sqrt(np.sin(t)))
    return y1
plt.plot(x_plot3, f03(x_plot3), linestyle="--",label="Analytical Solution")
plt.plot(x_plot3, y_plot3, label= "Solution from Scipy")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution of the 3rd equation')
plt.legend()
plt.grid(True)
plt.show()
```

## Equation 4

```python
def fun_4(x, y):
    return np.vstack((y[1], 0.5-(((y[1])**2)/2)-y[0]*np.sin(x)/2))

def bc_4(ya, yb):
    return np.array([ya[0]-2, yb[0]-2])

x_4 = np.linspace(0, 3.14, 5)
y_4 = np.zeros((2, x_4.size))

sol4 = solve_bvp(fun_4, bc_4, x_4, y_4)

x_plot4 = np.linspace(0, 3.14, 100)
y_plot4 = sol4.sol(x_plot4)[0]

def f04(t):
    y1=(2+(np.sin(t)))
    return y1
plt.plot(x_plot4, f04(x_plot4), linestyle="--",label="Analytical Solution")
plt.plot(x_plot4, y_plot4, label="Solution from Scipy")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution of the 4th equation')
plt.legend()
plt.grid(True)
plt.show()
```
