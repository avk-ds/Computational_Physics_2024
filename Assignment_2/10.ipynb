```python
import numpy as np
import matplotlib.pyplot as plt

def runge_kutta_4th_order(function, time, value, step):
    k1 = step * function(time, value)
    k2 = step * function(time + step / 2, value + k1 / 2)
    k3 = step * function(time + step / 2, value + k2 / 2)
    k4 = step * function(time + step, value + k3)
    return value + (k1 + 2 * k2 + 2 * k3 + k4) / 6

def differential_eq(time, value):
    return (value**2 + value) / time

min_time = 1
max_time = 3
step_size = 0.001
values = [-2]
times = [min_time]
a = 0

while a == 0:
    a = 2
    if times[-1] + step_size < max_time:
        a -= 1
        new_value = runge_kutta_4th_order(differential_eq, times[-1], values[-1], step_size)
        values.append(new_value)
        times.append(times[-1] + step_size)
        if times[-1] + step_size < max_time:
            new_value = runge_kutta_4th_order(differential_eq, times[-1], values[-1], step_size)
            a -= 1
            value_2_step = runge_kutta_4th_order(differential_eq, times[-2], values[-2], 2 * step_size)
            rho = (10 ** (-4) * step_size * 30 / np.abs(new_value - value_2_step)) ** (0.25)
            step_size *= rho

step_size = max_time - times[-1]
times.append(max_time)
values.append(runge_kutta_4th_order(differential_eq, times[-1], values[-1], step_size))

def true_function(t):
    return -2 * t / (-1 + 2 * t)

error = np.linalg.norm(values - np.vectorize(true_function)(times), ord=np.inf)

plt.plot(times, values, 'red')
for time in times:
    plt.axvline(x=time)

plt.xlabel('Time')
plt.ylabel('Value')
plt.title("Solution with adaptive step size in RK4")
plt.show()
```
