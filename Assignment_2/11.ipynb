```python
import numpy as np
import matplotlib.pyplot as plt
def f(x,t):
    return 1/(x**2 +t**2)
def rk4_step(f,x,t,h):
    k1=h*f(x,t)
    k2=h*f(x+k1/2,t+h/2)
    k3=h*f(x+k2/2,t+h/2)
    k4=h*f(x+k3,t+h)
    return (x+(k1+2*k2+2*k3+k4)/6)
def interpolated_value(t_given,t,x):
    indx=np.searchsorted(t,t_given)
    slp=(x[indx]-x[indx-1])/(t[indx]-t[indx-1])
    x_req=x[indx]+slp*(t_given-t[indx-1])
    return x_req
#Initialisation
x_old=1
x=[]
x.append(x_old)
t=np.linspace(0,np.power(10,8),10)
h=t[1]-t[0]
for i in t[:-1]:
    x_new=rk4_step(f,x_old,i,h)
    x.append(x_new)
    x_old=x_new
plt.plot(t,x)
plt.xlabel("t")
plt.ylabel("x")
plt.title("Solution by RK4")
plt.grid(True)
plt.plot()
plt.show()
#Finding the value at specified t:
x_1=interpolated_value(3.5*np.power(10,6),t,x)
print("The required value is ", x_1)
```
