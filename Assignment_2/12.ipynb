```python
import numpy as np
import matplotlib.pyplot as plt
def f_1(u_1,u_2,u_3,t):
    return (u_1+2*u_2-2*u_3+np.exp(-t))
def f_2(u_1,u_2,u_3,t):
    return (u_2+u_3-2*np.exp(-t))
def f_3(u_1,u_2,u_3,t):
    return (u_1+2*u_2+np.exp(-t))
def rk4(u_1, u_2, u_3, t, h):
    k1=h*f_1(u_1,u_2,u_3,t)
    l1=h*f_2(u_1,u_2,u_3,t)
    m1=h*f_3(u_1,u_2,u_3,t)
    k2=h*f_1(u_1+k1/2,u_2+l1/2,u_3+m1/2,t+h/2)
    l2=h*f_2(u_1+k1/2,u_2+l1/2,u_3+m1/2,t+h/2)
    m2=h*f_3(u_1+k1/2,u_2+l1/2,u_3+m1/2,t+h/2)
    k3=h*f_1(u_1+k2/2, u_2+m2/2, u_3 + l2/2, t+h/2)
    l3=h*f_2(u_1+k2/2, u_2+m2/2, u_3 + l2/2, t+h/2)
    m3=h*f_3(u_1+k2/2, u_2+m2/2, u_3 + l2/2, t+h/2)
    k4=h*f_1(u_1+k3, u_2+l3, u_3+m3, t+h)
    l4=h*f_2(u_1+k3, u_2+l3, u_3+m3, t+h)
    m4=h*f_3(u_1+k3, u_2+l3, u_3+m3, t+h)
    n_1=u_1+(k1+2*k2+2*k3+k4)*(1/6)
    n_2=u_2+(l1+2*l2+2*l3+l4)*(1/6)
    n_3=u_3+(m1+2*m2+2*m3+m4)*(1/6)
    return n_1, n_2, n_3
t=np.linspace(0,1,100)
u_1=[]
u_2=[]
u_3=[]
u_1i=3
u_2i=-1
u_3i=1
u_1.append(u_1i)
u_2.append(u_2i)
u_3.append(u_3i)
h=t[1]-t[0]
for i in t[:-1]:
    x_1, x_2, x_3 = rk4(u_1i, u_2i, u_3i, i, h)
    u_1.append(x_1)
    u_2.append(x_2)
    u_3.append(x_3)
    u_1i=x_1
    u_2i=x_2
    u_3i=x_3
plt.plot(t, u_1, label="u_1")
plt.plot(t, u_2, label="u_2")
plt.plot(t, u_3, label="u_3")
plt.xlabel("t")
plt.ylabel("u")
plt.legend()
plt.grid(True)
plt.show()
print("The values of u_1 are:\n", u_1)
print("The values of u_2 are:\n", u_2)
print("The values of u_3 are:\n", u_3)
```
