Putting $dy/dt=v$ we will get,

$\frac{dv}{dt}=\frac{t^3ln(t)-2y+2vt}{t^2}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions f_1(t, v, y) and f_2(t, v, y)
def f_1(t, v, y):
    return (t**3 * np.log(t) - 2*y + 2*v*t) / t**2
def f_2(t, v, y):
    return v

# Implement Euler's method
def euler(f_1, f_2, t, v, y, h):
    v_new = v + h * f_1(t, v, y)
    y_new = y + h * f_2(t, v, y)
    return v_new, y_new

# Define step size and initial conditions
h = 0.001
t_start = 1
t_end = 2
y_initial = 1
v_initial = 0

# Initialize arrays to store solution
y_values = []
t_values = np.arange(t_start, t_end + h, h)
print(t_values)
# Perform Euler's method
y_old = y_initial
y_values.append(y_old)
v_old = v_initial
for t in t_values[:-1]:
    v_new, y_new = euler(f_1, f_2, t, v_old, y_old, h)
    y_values.append(y_new)
    v_old = v_new
    y_old = y_new

# Plot the solution
plt.plot(t_values, y_values, label="Solution from Euler\'s method")
plt.xlabel('t')
plt.ylabel('y')
plt.title('Solution using Euler\'s method')
plt.grid(True)

# Plotting the exact solution
def y_exact(t):
    return 7/4 * t + (t**3)/2 * np.log(t) - 3/4 * t**3

t_exact = np.linspace(t_start, t_end, 1000)
y_exact_values = y_exact(t_exact)
plt.plot(t_exact, y_exact_values, linestyle="--", label="Exact solution")

plt.legend()
plt.show()
```
