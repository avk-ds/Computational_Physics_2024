## Euler method

```python
import numpy as np
import matplotlib.pyplot as plt
def f(y,t):
    z=(y/t)-(y/t)**2
    return z
h=0.1
t_array=np.arange(1,2+h,h)
y0=1
y_array=[]
y_array.append(y0)
for i in t_array[:-1]:
    y_1=y0+h*f(y0,i)
    y_array.append(y_1)
    y0=y_1
plt.plot(t_array, y_array)
plt.xlabel('t')
plt.ylabel('y')
plt.title("Solution from Euler's Method")
plt.show()
```

## Analytical solution

```python
def g(t):
    z_1=t/(1+np.log(t))
    return z_1
soln_y_array=[]
y_old=1
for j in t_array:
    soln_y_array.append(g(j))
plt.plot(t_array, soln_y_array)
plt.xlabel("t")
plt.ylabel("y")
plt.title("Analytical Solution")
plt.show()
```

## Finding the absolute errors:

```python
def abs_error(true_soln, approx_soln):
    z_1=np.absolute(np.array(true_soln)-np.array(approx_soln))
    return z_1
print("The absolute error is ", abs_error(soln_y_array, y_array))
plt.plot(t_array, abs_error(soln_y_array, y_array))
plt.xlabel('t')
plt.ylabel('Absolute error')
plt.title("Absolute Errors")
```

## Finding the relative error

```python
def rel_error(true_soln, approx_soln):
    z_1=np.absolute((np.array(true_soln)-np.array(approx_soln))/(np.array(true_soln)))
    return z_1
print("The relative error is ", rel_error(soln_y_array, y_array))
plt.plot(t_array, rel_error(soln_y_array, y_array))
plt.xlabel('t')
plt.ylabel('Relative error')
plt.title("Relative Errors")
```
