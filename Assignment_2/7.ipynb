```python
import numpy as np
import matplotlib.pyplot as plt

t_array = np.linspace(0, 10, 100)
h = t_array[1] - t_array[0]
x_array1 = np.zeros(100)
x_array2 = np.zeros(100)
x_array1[0] = 0  # Boundary condition at t=0
x_array1[-1] = 0  # Boundary condition at t=t1
g = 10
tol = 0.001
max_iter = 1000
for _ in range(max_iter):
    for i in range(1, len(x_array1) - 1):
        x_array1[i] = (1/2) * (x_array1[i + 1] + x_array1[i - 1] + g * h**2)
    diff = np.abs(x_array2 - x_array1)
    if np.linalg.norm(diff) <= tol:
        break
    else:
        np.copyto(x_array2, x_array1)
plt.plot(t_array, x_array1)
plt.xlabel("t")
plt.ylabel("x")
plt.title("Relaxation method")
plt.show()
```
