```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from file
random_data = np.loadtxt('exponential_distribution.txt')

# Plot the histogram of the data
plt.hist(random_data, bins=100, density=True, label='Generated Data')

# Parameters for the exponential distribution
rate_param = 2.0  
x_values = np.linspace(0, np.max(random_data), 1000)
pdf_values = rate_param * np.exp(-rate_param * x_values)

# Plot the exact PDF
plt.plot(x_values, pdf_values, 'r-', lw=2, label='Exact PDF')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('P(x)')
plt.legend()

# Show the plot
plt.show()
```
