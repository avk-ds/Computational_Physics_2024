```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_distribution(value, mean, std_dev):
    return np.exp(-(value - mean)**2 / (2 * std_dev**2)) / (2 * np.pi * std_dev**2)**0.5

sample_size = 10000
x_samples, y_samples = [], []
for i in range(sample_size):
    rand_val1, rand_val2 = np.random.rand(), np.random.rand()
    radius = (-2 * np.log(rand_val1))**0.5
    angle = 2 * np.pi * rand_val2
    x_samples.append(radius * np.cos(angle))
    y_samples.append(radius * np.sin(angle))

x_values = np.linspace(-5, 5, 100)

figure, (axis1, axis2) = plt.subplots(1, 2, figsize=(10, 6))
axis1.hist(x_samples, 50, density=True, label='Numerical')
axis1.set_xlabel('x')
axis1.set_ylabel('P(x)')
axis1.plot(x_values, gaussian_distribution(x_values, 0, 1), label='Exact')
axis1.legend()

axis2.hist(y_samples, 50, density=True, label='Numerical')
axis2.set_xlabel('y')
axis2.set_ylabel('P(y)')
axis2.plot(x_values, gaussian_distribution(x_values, 0, 1), label='Exact')
axis2.legend()

plt.show()
```
