```python
import numpy as np
import matplotlib.pyplot as plt

def target_distribution(x):
    return (2/np.pi)**0.5 * np.exp(-x**2/2)

num_samples = 100000
random_samples = []

for j in range(num_samples):
    x_sample = 10 * np.random.rand()
    y_sample = (2/np.pi)**0.5 * np.random.rand()
    if y_sample < target_distribution(x_sample):
        random_samples.append(x_sample)

plt.hist(random_samples, bins=100, density=True, label='Rejection method')
plt.plot(np.linspace(0, 5, 100), target_distribution(np.linspace(0, 5, 100)), label='Exact distribution')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
