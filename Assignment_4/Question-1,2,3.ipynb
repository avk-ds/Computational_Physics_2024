# Question 1

```python
import time
import matplotlib.pyplot as plt
import numpy as np

# Parameters for the Linear Congruential Generator
a, c, m = 214013, 2531011, 2**31
seed = 12345
random_numbers = []

# Measure start time
start_time = time.time_ns()

# Generate 10,000 random numbers
for i in range(10000):
    seed = (a * seed + c) % m
    random_numbers.append(seed / m)

# Measure end time
end_time = time.time_ns()

# Plot the histogram of the generated random numbers
plt.hist(random_numbers, 100, density=True, label='Linear Congruential')
plt.plot([0, 1], [1,1], label='Uniform')
plt.xlim([0, 1])
plt.ylim([0, 1.5])
plt.legend()
plt.show()
```

# Question 2

```python

# Initialize an empty list to store the generated random numbers
random_values = []
start_time_np = time.time_ns()

# Generate 10,000 random numbers using NumPy
for j in range(10000):
    random_values.append(np.random.rand())

end_time_np = time.time_ns()

# Plot the histogram of the generated random numbers
plt.hist(random_values, 100, density=True, label='Numpy')
plt.plot([0, 1], np.ones(2), label='Uniform')
plt.xlim([0, 1])
plt.ylim([0, 1.5])
plt.legend()
plt.show()
```

# Question 3

```python
print("The time taken to generate 10,000 random numbers using Linear Congruential Generator is", (end_time - start_time), "nano-seconds")
print("The time taken to generate 10,000 random numbers using NumPy is", (end_time_np - start_time_np), "nano-seconds")
```
